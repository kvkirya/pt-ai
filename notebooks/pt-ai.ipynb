```python
#I am going to load up the file into my jupyter notebook
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import json

with open('annotations.json') as json_data:
    data = json.load(json_data)
```

```python
df_images = pd.DataFrame(data["images"])
df_annotations = pd.DataFrame(data["annotations"])
```

```python
data_annotations = data["annotations"]
```

```python
data_annotations_squats = []
for i in data_annotations:
    if i["pose_variation"] == "squats":
        data_annotations_squats.append(i)
```

```python
df_squats_annotations = pd.DataFrame(data_annotations_squats)
df_squats_annotations.drop(columns = "id", axis = 0, inplace = True)
df_squats_annotations.rename(columns = {"image_id" : "id"}, inplace = True)
df_squats_annotations
```

```python
df_images
```

```python
merged_squats_df = pd.merge(df_squats_annotations, df_images, on = "id")
merged_squats_filtered_df = merged_squats_df[['id', 'file_name', 'pose_variation',
       'pose_category', 'exercise_category']]
merged_squats_filtered_df
```

```python
merged_squats_df.keys()
```

```python
filename = merged_squats_filtered_df['file_name']
filenames =filename.to_list()
```

```python
clean_files = []
for file in filenames:
    if file.endswith('.rgb.png'):
        clean_files.append(file)
```

```python
import os
import shutil
```

```python
source_folder = '00'
```

```python
new_folder_name = "00_squats"
```
